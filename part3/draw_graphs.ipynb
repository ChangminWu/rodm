```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
file = "email-Eu-dist.txt"
degs = []
dists = []
with open(file, "r") as f:
    for line in f.readlines():
        degs.append(int(line.split(" ")[0]))
        dists.append(int(line.split(" ")[1]))
```

```python
plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(15, 15))
plt.ylim([0, 20])
plt.xlim([0, max(degs)])
plt.bar(degs, height= dists, width=10)
plt.savefig("Eu-partial.eps", dpi=1200, format="eps", transparent=True, bbox_inches="tight", pad_inches=0.1)
```
